```python
def remove_dups(folderpath): # Remove Duplicate songs
    print(os.getcwd())
    os.chdir(folderpath)
    songlist = defaultdict()
    for album in os.listdir():
        os.chdir(album)
        print(os.getcwd())
        for song in os.listdir():
            
            songpath = os.getcwd() + '/' + song
            songname = song.replace(" ", "").lower()
            if songname in songlist:
                print("removing " + songname)
                os.unlink(songpath)
            else:
                songlist[songname] = 1
        os.chdir('..')
    os.chdir(code_path)
```

```python
def extract_notes(filepath):
    # Convert the midi files into strings, and put them in a list
    os.chdir(filepath)
    notes = []
    for album in os.listdir():
        os.chdir(album)
        print(os.getcwd())
        for song in os.listdir():
            print("Converting Song " + song)
            try:
                s = converter.parse(song)
            except:
                continue
            songNotes = None
            try:
                instruments = instrument.partitionByInstrument(s)
                songNotes = s.parts[0].recurse()
            except:
                songNotes = s.flat.notes
            for element in songNotes:
                if isinstance(element, note.Note):
                    notes.append(str(element.pitch))
                elif isinstance(element, chord.Chord):
                    notes.append('.'.join(str(n) for n in element.normalOrder))
        os.chdir('..')
    os.chdir('..')
    return notes
```

```python
def create_dicts(pitchnames):
    pitch_to_int = dict((pitch, number) for number, pitch in enumerate(pitchnames))
    int_to_pitch = dict((number, pitch) for number, pitch in enumerate(pitchnames))
    return pitch_to_int, int_to_pitch

def create_sequence_matrices(seqlen, pitchnames, notes, pitch_to_int):
    step = seqlen
    sequences = []
    for i in range(0, len(notes) - seqlen - 1, step):
        sequences.append(notes[i: i + seqlen + 1])
    x = np.zeros((len(sequences), seqlen, len(pitchnames)), dtype=np.bool)
    y = np.zeros((len(sequences), seqlen, len(pitchnames)), dtype=np.bool)
    for i, sequence in enumerate(sequences):
        for t, (note_in, note_out) in enumerate(zip(sequence[:-1], sequence[1:])):
            x[i, t, pitch_to_int[note_in]] = 1
            y[i, t, pitch_to_int[note_out]] = 1
    return x, y


def create_model(seqlen, pitchnames):
    model = Sequential()
    model.add(LSTM(512, input_shape=(seqlen, len(pitchnames)), return_sequences=True))
    model.add(LSTM(512, return_sequences=True, recurrent_dropout=0.3,))
    model.add(BatchNorm())
    model.add(Dropout(0.3))
    model.add(Dense(256))
    model.add(Activation('relu'))
    model.add(BatchNorm())
    model.add(Dropout(0.3))
    model.add(Dense(len(pitchnames)))
    model.add(Activation('softmax'))
    model.compile(
      loss='categorical_crossentropy',
      optimizer=RMSprop(learning_rate=0.01),
      metrics=['categorical_crossentropy', 'accuracy']
    )
    return model

def sample(preds, temperature=1.0):
    """Helper function to sample an index from a probability array."""
    preds = np.asarray(preds).astype('float64')
    preds = np.exp(np.log(preds) / temperature)  # softmax
    preds = preds / np.sum(preds)                #
    probas = np.random.multinomial(1, preds, 1)  # sample index
    return np.argmax(probas)                     #


def on_epoch_end(epoch, _):
    """Function invoked at end of each epoch. Prints generated text."""
    print()
    print('----- Generating text after Epoch: %d' % epoch)
    start_index = random.randint(0, len(notes) - seqlen - 1)
    for diversity in [.5]:
        print('----- diversity:', diversity)

        generated = ''
        sequence = notes[start_index: start_index + seqlen]
        sequence = ', '.join(sequence)
        generated += sequence
        print('----- Generating with seed: "' + sequence + '"')
        sys.stdout.write(generated)

        for i in range(400):
            x_pred = np.zeros((1, seqlen, len(pitchnames)))
            for t, n in enumerate(sequence.split(", ")):
                x_pred[0, t, pitch_to_int[n]] = 1.
                preds = model.predict(x_pred, verbose=0)
            
            next_index = sample(preds[0, -1], diversity)
            next_note = int_to_pitch[next_index]
            sequence = sequence.split(', ')[1:]
            sequence.append(next_note)
            sequence = ', '.join(sequence)
            sys.stdout.write(", " + next_note)
            sys.stdout.flush()
        print()

filepath = "weights-improvement-{epoch:02d}-{loss:.4f}-bigger.hdf5"
checkpoint = ModelCheckpoint(
    filepath,
    monitor='loss',
    verbose=0,
    save_best_only=True,
    mode='min'
)
```

```python
model_take_1 = create_model(seqlen, pitchnames_take_1)

notes = notes_take_1
pitchnames = pitchnames_take_1
pitch_to_int = pitch_to_int_take_1
model = model_take_1
int_to_pitch = int_to_pitch_take_1
print_callback = LambdaCallback(on_epoch_end=on_epoch_end)

model_take_1.fit(x_take_1, y_take_1,
          batch_size=128,
          epochs=100,
          callbacks=[print_callback])
```

```python
def sample(preds, temperature=1.0):
    """Helper function to sample an index from a probability array."""
    preds = np.asarray(preds).astype('float64')
    preds = np.exp(np.log(preds) / temperature)  # softmax
    preds = preds / np.sum(preds)                #
    probas = np.random.multinomial(1, preds, 1)  # sample index
    return np.argmax(probas)   

def generate_song(notes, model, name, pitchnames, pitch_to_int, int_to_pitch):
    generated = ''
    sequence = notes[100:140]
    whole = sequence
    sequence = ', '.join(sequence)
    generated += sequence
    print('----- Generating with seed: "' + sequence + '"')
    sys.stdout.write(generated)

    for i in range(200):
        x_pred = np.zeros((1, seqlen, len(pitchnames)))
        for t, n in enumerate(sequence.split(", ")):
            x_pred[0, t, pitch_to_int[n]] = 1.
        preds = model.predict(x_pred, verbose=0)

        next_index = sample(preds[0, -1], .5)
        next_note = int_to_pitch[next_index]
        whole.append(next_note)
        sequence = sequence.split(', ')[1:]
        sequence.append(next_note)
        sequence = ', '.join(sequence)

        sys.stdout.write(", " + next_note)
        sys.stdout.flush()
    print()
    print("whole: ", whole)
    offset = 0
    output_notes = []

  # create note and chord objects based on the values generated by the model
    for pattern in whole:
      # pattern is a chord
        if ('.' in pattern) or pattern.isdigit():
          # if pattern == '0':
          #   continue
            notes_in_chord = pattern.split('.')
            notes = []
            for current_note in notes_in_chord:
                new_note = note.Note(int(current_note))
                new_note.storedInstrument = instrument.Guitar()
                notes.append(new_note)
            new_chord = chord.Chord(notes)
            new_chord.offset = offset
            output_notes.append(new_chord)
        else:
            new_note = note.Note(pattern)
            new_note.offset = offset
            new_note.storedInstrument = instrument.Guitar()
            output_notes.append(new_note)
        offset += 0.5

    midi_stream = stream.Stream(output_notes)
    midi_stream.show('midi')
    midi_stream.write('midi', fp='../Song Attempts/'+name)
    print("done")
    

def generate_actual_notes(notes, name, pitchnames, pitch_to_int, int_to_pitch):
    generated = ''
    whole = notes[100:300]
    offset = 0
    output_notes = []

  # create note and chord objects based on the values generated by the model
    for pattern in whole:
      # pattern is a chord
        if ('.' in pattern) or pattern.isdigit():
          # if pattern == '0':
          #   continue
            notes_in_chord = pattern.split('.')
            notes = []
            for current_note in notes_in_chord:
                new_note = note.Note(int(current_note))
                new_note.storedInstrument = instrument.Guitar()
                notes.append(new_note)
            new_chord = chord.Chord(notes)
            new_chord.offset = offset
            output_notes.append(new_chord)
        else:
            new_note = note.Note(pattern)
            new_note.offset = offset
            new_note.storedInstrument = instrument.Guitar()
            output_notes.append(new_note)
        offset += 0.5

    midi_stream = stream.Stream(output_notes)
    midi_stream.show('midi')

    midi_stream.write('midi', fp='../Song Attempts/'+name)
    print("done")
```
