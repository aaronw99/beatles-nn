```python
import requests
from bs4 import BeautifulSoup
import urllib.request

import os
from os import getcwd
path = "C:\\Users\\Maciej Zielonka\\Documents\\Beatles"

os.chdir(path)
```

```python
homepage = "http://bmh.webzdarma.cz/midieng.htm?fbclid=IwAR2oc4BEnzd_OOa_VViRz3DVDTijPvpFqSqb26151P5Vr_RlLdbTdX8Dolk"
web = "http://bmh.webzdarma.cz/"
```

```python
r = requests.get(homepage)
s = BeautifulSoup(r.text)
allhrefs = s.findAll("a", href=True)
# print(allhrefs['text'])
albumnames = []
albumslinks = []
for l in allhrefs:
    albumnames.append(l.text)
    albumslinks.append(l['href'])
print(albumnames)
```

```python
for name, album in zip(albumnames, albumslinks):
#     print(name)
    if name == 'Singles/ "1"':
        name = 'Singles'
    newfolder = path + '\\' + name
    if not os.path.isdir(newfolder):
        print("Making folder " + newfolder)
        os.mkdir(newfolder, 777)
    print("Going into " + newfolder)
    os.chdir(newfolder)
    page = web + album
    pager = requests.get(page)
    s = BeautifulSoup(pager.text)
    songhrefs = s.findAll("a", href=True)
    songnames = []
    songlinks = []
    for l in songhrefs:
        songnames.append(l.text)
        songlinks.append(l['href'])
    songsadded = []
    for song, songname in zip(songlinks, songnames):
        
        specificlink = web + song
        if song == 'index.htm':
            break
        try:
            if not os.path.isfile(newfolder + '\\' + songname + ".mid") and not (songname in songsadded):
                print("downloading song " + songname)
                songsadded.append(song)
                urllib.request.urlretrieve(specificlink, songname + ".mid")
            elif song in songsadded:
                print("already added " + song)
        except:
            continue
#     print(songlinks)
    os.chdir(path)
    print("back in parent")
    
#     break
```
